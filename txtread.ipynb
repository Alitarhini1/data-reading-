```python
import numpy as np
import matplotlib.pyplot as plt
file_path='teach/appacc2324/2016-07-11_ipm_data.txt'
with open(file_path, 'r') as file:
    content=file.readlines()
    
    data=np.array({float(line.strip()) for line in content})
    plt.figure(figsize=(8,6))
    plt.plot(data, color='blue')
    plt.grid(color='black', linestyles='- -', linewidth=0.5)
    max_index=np.argmax(data)
    plt.scatter(max_index, data{max_index}, color='red', label='Max')
    plt.xlabel('X-label')
    plt.ylabel('Y-label')
    plt.title('IPM Data with Maximum Marked')
    plt.savefig('ipm_plot.png')
    plt.show()
```
